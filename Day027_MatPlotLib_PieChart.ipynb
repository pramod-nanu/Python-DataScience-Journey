```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 35, 10]

plt.pie(sizes, labels=labels)
plt.title("Fruit Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
bowler = ['Bumrah','Siraj','Jadeja','Akashdeep','Washington']
wickets = [3,2,2,2,12]
plt.pie(wickets,labels=bowler,autopct = '%1.1f%%')
plt.title("Wickets Pie-Chart")
plt.show()


```

```python
### explode
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]
explode = [0.1, 0, 0, 0]  # Multiple slices exploded
  # Explode the 1st slice (Apples)

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',startangle=90,shadow=True)
plt.title("Fruit Distribution")
plt.axis('equal')  # Keeps it circular
plt.show()
```

```python
### Color
import matplotlib.pyplot as plt
labels = ['Kohli','Rohit','Rahul','Gill','Yashaswi']
runs = [10000,8000,4000,2000,1900]
colors = ['Red','green','blue','purple','orange']
explode = [0.1,0,0,0,0]
plt.pie(runs,labels=labels,colors=colors,explode = explode,startangle = 90,shadow=True,counterclock=False)
plt.show()
```

```python
### Donut Chart
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create donut
plt.pie(
    sizes,
    labels=labels,
    radius=1.2,  # Makes the entire donut bigger
    wedgeprops={'width': 0.4}
)
plt.title("Donut Chart with Custom Radius")
plt.axis('equal')
plt.show()

```
