### Data Distribution Basics

### Normal Distribution (Bell Curve)
Most data points cluster around the mean (center).

The distribution is symmetric.

Bell-shaped curve

Mean = Median = Mode

68–95–99.7 Rule (Empirical Rule):

68% of data lies within 1 standard deviation from the mean.

95% lies within 2 standard deviations.

99.7% lies within 3 standard deviations.
Real-life Examples

Heights of people in a large population.

IQ scores.

Measurement errors in experiments.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate normally distributed data
data = np.random.normal(loc=50, scale=10, size=1000)
# loc = mean, scale = std dev, size = sample size

# Plot histogram + KDE
sns.histplot(data, kde=True, color='blue', bins=30)
plt.axvline(np.mean(data), color='red', linestyle='--', label=f'Mean = {np.mean(data):.2f}')
plt.title("Normal Distribution Example")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
tips = sns.load_dataset('tips')

# Extract total_bill column
data = tips['total_bill']

# Plot histogram + KDE
sns.histplot(data, kde=True, color='green', bins=20)
plt.axvline(np.mean(data), color='red', linestyle='--', label=f'Mean = {np.mean(data):.2f}')
plt.title("Distribution of Total Bill")
plt.legend()
plt.show()

# Check summary stats
mean_val = np.mean(data)
median_val = np.median(data)
std_dev = np.std(data)
print(f"Mean: {mean_val:.2f}, Median: {median_val:.2f}, Std Dev: {std_dev:.2f}")
```

### SkewNess
Left Skew (Negative)
Tail on the left
Mean < Median

Right Skew (Positive)
Tail on the right
Mean > Median

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

tips = sns.load_dataset('tips')
data = tips['total_bill']

# Plot histogram with KDE
sns.histplot(data, kde=True, color='purple', bins=20)
plt.title("Skewness Example - Total Bill")
plt.show()

# Calculate skewness
Skewness_value = skew(data)
print("Skewness value is : ",Skewness_value)
print(f"Skewness Value is : {Skewness_value :.2f}")
# Interpretation
if skewness_value > 0:
    print("Right Skewed (Positive Skew)")
elif skewness_value < 0:
    print("Left Skewed (Negative Skew)")
else:
    print("Perfectly Symmetric")
```

### Kurtosis
Kurtosis measures the "tailedness" of a distribution — how much of the data lies in the tails compared to the normal distribution.

Mesokurtic (kurtosis ≈ 3)

Leptokurtic (kurtosis > 3)

Platykurtic (kurtosis < 3)

High kurtosis → greater chance of outliers

Low kurtosis → data more evenly spread, fewer outliers


```python
import pandas as pd
import numpy as np

data = pd.Series(np.random.normal(0, 1, 1000))  # Normal distribution
print("Kurtosis:", data.kurt())  # Pandas uses Fisher’s definition (subtracts 3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t, uniform

x = np.linspace(-5, 5, 1000)

# Mesokurtic - Normal Distribution
y_normal = norm.pdf(x, 0, 1)

# Leptokurtic - t-distribution with low df (fat tails)
y_lepto = t.pdf(x, df=3)  # Lower df → heavier tails

# Platykurtic - Uniform distribution (flatter)
y_platy = uniform.pdf(x, loc=-2, scale=4)  # from -2 to 2

plt.figure(figsize=(8,6))
plt.plot(x, y_normal, label="Mesokurtic (Normal, kurtosis=3)")
plt.plot(x, y_lepto, label="Leptokurtic (Fat tails, kurtosis>3)")
plt.plot(x, y_platy, label="Platykurtic (Flat, kurtosis<3)")

plt.legend()
plt.title("Kurtosis Types Comparison")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

### Visuals: Histograms, KDE plots, Boxplots

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(50, 10, 1000)  # mean=50, std=10

plt.hist(data, bins=20, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(data, fill=True)
plt.title("KDE Plot")
plt.show()
```

```python
sns.boxplot(x=data)
plt.title("Boxplot")
plt.show()
```

### All 3 Plots in single diagram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = np.random.normal(50,10,1000)
plt.figure(figsize=(15,4))
# Histogram
plt.subplot(1,3,1)
plt.hist(data,bins = 20,edgecolor= 'black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('frequency')

# Kds plot
plt.subplot(1,3,2)
sns.kdeplot(data,fill = True,color = 'pink')
plt.title('KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')

# BoxPlot
plt.subplot(1,3,3)
sns.boxplot(x = data)
plt.title('Box Plot')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()

```
