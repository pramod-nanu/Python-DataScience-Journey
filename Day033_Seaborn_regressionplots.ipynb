### RegPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.regplot(data=tips, x='total_bill', y='tip')
plt.title("Linear Regression: Total Bill vs Tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
sns.regplot(data=titanic,x='fare',y='age',
           line_kws ={'color' : 'green'},
            scatter_kws= {'color':'blue'}
           )
plt.title("Fare vs Age with Regression Line")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

### Lmplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
sns.lmplot(data = titanic,x ='age',y = 'fare',hue= 'sex',col='class',
                line_kws={'color':'green'},
                scatter_kws={'color':'orange'})
plt.show()
```

### Polynomial Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
sns.lmplot(
    data=titanic,
    x='age',
    y='fare',
    order=2, # Polynomial of degree 2 (quadratic)
    line_kws={'color': 'red'},
    scatter_kws={'alpha': 0.5}
)
plt.title("Polynomial Regression (Quadratic): Fare vs Age")
plt.show()
```

### Displot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset
titanic = load_dataset('titanic')
sns.displot(data=titanic, x='age', kde=True, bins=10, hue='class', multiple='stack')
plt.show()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

titanic = load_dataset('titanic')

sns.histplot(data=titanic, x='age', bins=20, kde=True)
plt.title("Age Distribution with KDE")
plt.show()
```

```python
### kdeplot()
sns.kdeplot(data=titanic, x='age', hue='sex', fill=True, common_norm=False, alpha=0.4)
plt.title("Age Distribution by Sex (KDE)")
plt.show()
```

```python
### kde plot
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

titanic = load_dataset("titanic")

# Plot fare vs age density
sns.kdeplot(data=titanic, x='age', y='fare',fill=True,cmap="Reds", thresh=0.05,levels=10)
plt.title("2D KDE Plot: Fare vs Age")
plt.show()
```

```python
### JointPlot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

titanic = load_dataset("titanic")

sns.jointplot(data=titanic, x='age', y='fare', kind='reg', color='g')
plt.show()
```
