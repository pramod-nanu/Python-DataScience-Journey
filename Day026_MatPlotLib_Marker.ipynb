### Marker Customization

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(
    x, y,
    marker='o',                # circular marker
    markersize=10,             # size of marker
    markerfacecolor='blue', # fill color
    markeredgecolor='red',    # border color
    markeredgewidth=2,        # border thickness
    color='piNk',             # line color
    linewidth=2
)

plt.title("Marker Customization")
plt.grid(True)
plt.show()

```

```python
### color
import matplotlib.pyplot as plt
%matplotlib inline 
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [30, 25, 20, 15, 10]

plt.plot(x, y1, color='red', label='Red Line')        # Named color
plt.plot(x, y2, color='#0000ff', label='#ffFFff')   # Hex color

plt.title("Color Variations")
plt.legend()
plt.grid(True)
plt.show()
```

### plt.style.use()

```python
import matplotlib.pyplot as plt
# print(plt.style.available)

# Use a predefined style
plt.style.use('ggplot')

# Sample plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('Styled Plot using ggplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
