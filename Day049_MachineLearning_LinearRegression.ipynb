### Introduction to ML & Linear Regression (Part 1)

What is Machine Learning (ML)?

Machine Learning is a subset of Artificial Intelligence (AI) that enables systems to learn patterns from data and improve performance without being explicitly programmed.

In traditional programming:

You give rules + data â†’ program gives output.

In ML:

You give data + output (examples) â†’ program learns rules (model).

ðŸ”¹ Types of Machine Learning

Supervised Learning

Data has input (X) and output/label (Y).

Model learns to map inputs â†’ outputs.

Example: Predicting house prices (input = size, location; output = price).

Algorithms: Linear Regression, Logistic Regression, Decision Trees, etc.

Unsupervised Learning

Data has only input (X), no labels (Y).

Model tries to find hidden patterns.

Example: Customer segmentation in marketing.

Algorithms: K-Means Clustering, PCA, Hierarchical Clustering.

Reinforcement Learning (RL)

Agent learns by interacting with environment.

Uses rewards & penalties to improve decisions.

Example: Self-driving cars, AlphaGo (game-playing AI).

### Linear Regression: The Basics
1. Definition

Linear Regression is the simplest supervised ML algorithm used for predicting a continuous value.
It tries to fit a straight line through the data that best represents the relationship between input (X) and output (Y).

2. Equation of a Line

The line is represented as:

y=mx+c

Where:

y = predicted output (dependent variable)

x = input feature (independent variable)

m = slope of the line (how steep the line is, i.e., how much y changes when x increases)

c = intercept (value of y when x = 0)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# House size (X) and price (Y)
X = np.array([1000, 1500, 2000, 2500]).reshape(-1, 1)   # independent variable
y = np.array([50, 70, 90, 110])   # dependent Variable


model = LinearRegression()
model.fit(X,y)
# Check Slope and Intercept
print("Slope is : ",model.coef_[0])
print("Intercept is : ",model.intercept_)

# Predict for 1800 sqft
predicted_price = model.predict([[1800]])
print("Predicted Price for 1800 sqft:", predicted_price[0])

plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.xlabel("House Size (sqft)")
plt.ylabel("Price (â‚¹ lakh)")
plt.legend()
plt.show()
```
