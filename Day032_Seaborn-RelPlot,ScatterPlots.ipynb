### Relational,scatter plots in details

```python
### scatterplot() with hue, style, size, palette
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

sns.scatterplot(
    data=titanic,
    x='age',
    y='fare',
    hue='sex',
    style='class',
    size='survived',
    palette='Set2'
)

plt.title("Age vs Fare with Multiple Encodings")
plt.show()
```

```python
sns.scatterplot(
    data=titanic,
    x='age',
    y='fare',
    size='survived',
    sizes=(20, 50) ,# Small to large dot range
     edgecolor='pink',
    linewidth=0.5
)
plt.show()
```

### Lineplots

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

# Basic yearly trend
sns.lineplot(data=flights, x='year', y='passengers', hue='month', ci=None,style='month',dashes=True,marker='o')
plt.title("Passenger Trends (No Confidence Interval)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
tips = sns.load_dataset("tips")
sns.lineplot(data=tips, x='day', y='total_bill')
plt.title("Total Bill by Day (with 95% CI)")
plt.show()
```

```python
sns.lineplot(data=tips, x='day', y='total_bill', ci=None)
plt.title("Total Bill by Day (NO CI)")
plt.show()
```

### ScatterPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col='day',       # Creates 1 subplot per day # It could be row as well
    kind="line"   # It could be scatter as well
)

plt.suptitle("Tip vs Total Bill - Faceted by Day", y=1.05)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")
sns.regplot(
    data=tips,
    x='total_bill',
    y='tip',
    ci=None,
    scatter_kws={'color': 'blue', 'alpha': 0.6},
    line_kws={'color': 'Yellow', 'linewidth': 2}
)


plt.title("Regression Line: Total Bill vs Tip")
plt.show()
```

### Grouped Regression with sns.lmplot()


```python
sns.lmplot(data=tips, x='total_bill', y='tip',col='smoker', hue='sex')
plt.suptitle("lmplot: Tip vs Total Bill by Gender", y=1.02)
plt.show()
```

```python
sns.regplot(
    data=tips,
    x='total_bill',
    y='tip',
    line_kws={'color': 'red', 'linewidth': 2, 'linestyle': '--'}
)
```
