### Topic 1 : Top run scorers 

```python
# Step 1: Load the aggregated dataset
import pandas as pd
df = pd.read_csv("Data/IPL_Playerwise_Data.csv")
print(df.head())
print("#############")
print("#############")

# Step 2: Sort players by total runs
top_scorers = df.sort_values(by="TotalRuns", ascending=False).head(10)
print(top_scorers[["Player", "TotalRuns"]])
print("#############")
print("#############")

# Visualize
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.barplot(y='Player',x='TotalRuns',data =top_scorers,palette="Blues_r")
plt.title("Top Run Scorer")
plt.xlabel('Total Runs')
plt.ylabel('Players Name')
plt.show()
```

### Topic 2: Top Strike Rates (with min 100 balls faced)

```python
print(df.head())

# Step 2: Filter players (â‰¥ 100 balls)
qualified = df[df['TotalBalls'] >= 100]
print("#############")
print("#############")

# Step 3: Sort by Strike Rate
top_strike = qualified.sort_values(by='StrikeRate',ascending = False).head(10)
print(top_strike[['Player','StrikeRate','TotalRuns','TotalBalls']])
print("#############")
print("#############")

# Step 4: Visualize (Bar Chart)
sns.barplot(x='StrikeRate',y='Player',data=top_strike,palette="Greens_r")
plt.title("Best StrikeRate")
plt.xlabel('Player')
plt.ylabel('StrikeRate')
plt.plot()
```

### Topic 3: Boundary %

```python
df.head(5)
df['BoundryPCT'] = ((df['Fours'] * 4 + df['Sixes'] * 6) / df['TotalRuns']) * 100
df['BoundryPCT']= df['BoundryPCT'].round(1)

# Step 3: Sort by boundary %
top_boundry = df.sort_values(by = 'BoundryPCT',ascending= False).head(5)
print(top_boundry[['Player','TotalRuns','BoundryPCT']])

# Step 4: Visualize
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x = 'BoundryPCT',y = 'Player',data = top_boundry,palette='Oranges_r')
plt.xlabel("Boundary % of Runs")
plt.ylabel("Player")
plt.show()
```

### Topic 4: Bar chart for Top 10 Run Scorers

```python
# Step 1: Get Top 10 Run Scorers
top_scorers = df.sort_values(by ='TotalRuns',ascending = False).head(10)
#print(top_scorers[['Player','TotalRuns']])
plt.figure(figsize=(10,5))
sns.barplot(x="TotalRuns", y="Player", data=top_scorers, palette="Blues_r")
plt.title("Top 10 Run Scorers")
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.show()
```

### Topic 5: Scatterplot â€“ Runs vs Strike Rate

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x="TotalRuns", y="StrikeRate", data=df, hue="Player", s=100)
plt.title("Runs vs Strike Rate")
plt.xlabel("Total Runs")
plt.ylabel("Strike Rate")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")  # Move legend outside
plt.show()
```
