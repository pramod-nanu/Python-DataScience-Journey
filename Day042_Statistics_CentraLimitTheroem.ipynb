### Central Limit Theorem & Sampling

### Central Limit Theorem
When you average random samples, extreme values cancel out each other.

The result drifts towards the center (mean) of the population.

Repeating this many times → the distribution of these averages smooths into a bell curve.

### Mathematical Insights

Suppose population has mean μ and variance σ².

If we take a sample of size n:

The expected value of the sample mean = μ

The variance of the sample mean = σ² / n

The standard error (SE) = σ / √n

As n increases, SE decreases → averages are tightly clustered around μ.

```python
import numpy as np
import matplotlib.pyplot as plt

# Population: Dice (1 to 6)
population = [1, 2, 3, 4, 5, 6]

# Function to simulate sample means
def sample_means(sample_size, num_samples=10000):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        means.append(np.mean(sample))
    return means

# Different sample sizes
sample_sizes = [1, 2, 10, 30,1000,10000]

for n in sample_sizes:
    means = sample_means(n)
    plt.hist(means, bins=30, density=True, color='skyblue', edgecolor='black')
    plt.title(f"Sampling Distribution of Dice Means (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency (Density)")
    plt.show()
```

### Types Of Sampling

Visual Analogy

Imagine a jar with red and blue balls (population):

Random sampling: Pick balls completely at random.

Stratified: Separate reds and blues → then sample proportionally.

Systematic: Pick every 5th ball.

Cluster: Divide jar into 10 sections, pick 2 sections → take all balls from them.

Convenience: Just grab the top layer.

```python
import numpy as np
import pandas as pd

# Create a population dataset
np.random.seed(42)
population_size = 1000
population = pd.DataFrame({
    "id": range(1, population_size + 1),
    "age": np.random.randint(18, 26, population_size),
    "gender": np.random.choice(["M", "F"], population_size)
})

print("Population shape:", population.shape)
print(population.head())
```

```python
# Take 50 random students
simple_random = population.sample(n=50, random_state=42)
print("Simple Random Sample:\n", simple_random.head())
```
