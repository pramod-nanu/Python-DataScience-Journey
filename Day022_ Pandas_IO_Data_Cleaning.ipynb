### Missing values for short table

```python
import pandas as pd
# Load the sample CSV
df = pd.read_csv('sample_missing.csv')
# Display the full data
print("ðŸ”¹ Original Data:\n", df)

# Find Missing Values
print("\nðŸ”Ž Missing value count in each column:")
print(df.isnull().sum())


# Find Non-Missing Values
print("\nðŸ”Ž Non-Missing value count in each column:")
print(df.notnull().sum())

# Remove Rows with Any Missing Value
df_dropped = df.dropna()
print("\n After dropping rows with missing values:")
print(df_dropped)

# Fill Missing Age with Mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("\nðŸ“Œ Filled missing Age with mean:")
print(df)

# Fill Gender with Mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print("\nðŸ“Œ Filled missing Gender with mode:")
print(df)

# Fill Department with 'Unknown'
df['Department'] = df['Department'].fillna('Unknown')
print("\nðŸ“Œ Filled missing Department with 'Unknown':")
print(df)

#  Fill Salary with Median
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
print("\nðŸ“Œ Filled missing Salary with median:")
print(df)


```

### Topic 3: Data Type Conversion

```python
df = pd.read_csv('SampleDataTypeConverison.csv')
print("ðŸ”¹ Original Data:\n", df)

#  Check Current Data Types
print("\nðŸ”¹ Current Data Types :",df.dtypes)

# Convert to Numeric
df_ConvertNumeric = df['Salary'] = pd.to_numeric(df['Salary'])
print("\nðŸ”¹ Converted into Numerics:")
print(df_ConvertNumeric)

# Convert to Date
df_ConvertDate = df['JoinDate'] = pd.to_datetime(df['JoinDate'],errors = 'coerce')
print("\nðŸ”¹ Converted into Date ")
print(df_ConvertDate)

# Convert to Category (for Categorical Data)
df_ConvertCategory = df['Gender'].astype('category')
print("\nðŸ”¹ Converted into Category ")
print(df_ConvertCategory)
```

### Renaming Columns in Pandas


```python
# Rename Specific Columns
print("Original Table is :\n",df)
print("\nRenaming Sepcific Columns : ")
df.rename(columns = {'Naav': 'Name'},inplace=True)
print(df)

# Rename All Columns at Once
print("\nRenaming All Columns at once: ")
df.columns = ['Name 1', 'Age 1', 'Gender 1', 'Department 1', 'Salary 1','Joining Date1']
print(df)

#  Clean All Column Names
# Letâ€™s remove spaces and convert all column names to lowercase
print("\nRemoving all column and lowering name once: ")
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
print(df)

print("\n Renameing all column again")
df.columns = ['Name','Age','Gender','Department','Salary','JoiningDate']
print(df)

# Saving Cleaned Data with to_csv()
# Fill Missing Age with Mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
print("\nðŸ“Œ Filled missing Age with mean:")
print(df)
# Fill for Dept
df['Department'] = df['Department'].fillna("Unknown")
print("\n",df)
# Fill for Salary
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\n",df)

df.to_csv('cleaned_data.csv', index=False)

```
