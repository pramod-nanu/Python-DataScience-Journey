At each step, it asks a yes/no question about the data ‚Üí splits into groups ‚Üí until it can make a decision.

### Splitting

At each node (question), the tree splits the data into groups.

The goal: make groups as pure as possible (each group should mostly belong to one class).


### Gini Impurity (measure of ‚Äúimpurity‚Äù)

Think of it as: ‚ÄúHow mixed is this group?‚Äù

If a group has all same class ‚Üí Gini = 0 (pure).

If a group is half-half (like 50% yes, 50% no) ‚Üí Gini is higher (impure).

```python
# Step 1: Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 2: Create the dataset
data = {
    "Age": [12, 35, 40, 10],
    "Weather": ["Hot", "Hot", "Cold", "Cold"],
    "Buy": ["Yes", "Yes", "No", "No"]
}

df = pd.DataFrame(data)

# Step 3: Convert categorical column "Weather" into numbers
df["Weather"] = df["Weather"].map({"Hot": 1, "Cold": 0})
df["Buy"] = df["Buy"].map({"Yes": 1, "No": 0})

X = df[["Age", "Weather"]]   # Features
y = df["Buy"]                # Target

# Step 4: Create and train the decision tree
model = DecisionTreeClassifier(criterion="gini", random_state=42)
model.fit(X, y)

# Step 5: Visualize the decision tree
plt.figure(figsize=(6,4))
plot_tree(model, feature_names=["Age", "Weather"], class_names=["No", "Yes"], filled=True)
plt.show()

# Step 6: Make predictions
print("Prediction for Age=15, Weather=Hot:", model.predict([[15, 1]])[0])  # Expect Yes
print("Prediction for Age=30, Weather=Cold:", model.predict([[30, 0]])[0]) # Expect No
```
