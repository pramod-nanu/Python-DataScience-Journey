### Bar Charts with Matplotlib

### Simple Vertical Bar Chart

```python
import matplotlib.pyplot as plt
countries = ['India','USA','UK','Bangladesh','China']
population = [1400,900,500,800,1200]
plt.barh(countries,population,color='green')
plt.title('Countrywise Population',color='blue')
plt.xlabel('Country',color='blue')
plt.ylabel('Population',color='blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data
categories = ['India', 'USA', 'UK', 'Germany']
values = [120, 80, 60, 100]
plt.barh(categories, values)
plt.title("Country-wise Value Comparison (Horizontal)")
plt.xlabel("Values")
plt.ylabel("Country")
plt.show()
```

```python
## Custom Colors,Controlling Bar Width and Edge Color
categories = ['India', 'USA', 'UK', 'Germany']
values = [120, 80, 60, 100]
color = ['blue','green','yellow','pink']
plt.bar(categories,values,color = color,width=0.7,edgecolor= 'blue',linewidth=2)
plt.show()
```

### Multiple Bar charts

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['Q1','Q2','Q3','Q4']
Men = [20,30,35,30]
Women = [15,35,30,50]
x = np.arange(len(labels))
width = 0.35
plt.bar(x-width/2,Men,width,label = 'Men',color = 'blue')
plt.bar(x+width/2,Women,width,label = 'Women',color = 'pink')
plt.title('Gender Wise Performance')
plt.xticks(x,labels)
plt.ylabel('Scores')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Q1', 'Q2', 'Q3', 'Q4']
product_A = [20, 35, 30, 35]
product_B = [25, 32, 34, 20]

x = np.arange(len(labels))  # positions of bars

plt.bar(x, product_A, label='Product A', color='skyblue')
plt.bar(x, product_B, bottom=product_A, label='Product B', color='orange')

plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Quarterly Sales Comparison (Stacked)')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Q1', 'Q2', 'Q3', 'Q4']
Men = [20, 30, 35, 30]
x = np.arange(len(labels))

bars = plt.bar(x, Men, color='green')

plt.xticks(x, labels)
plt.ylabel('Scores')
plt.title('Men Performance with Annotations')

# Add annotations
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, yval, ha='center', va='bottom')

plt.show()
```
