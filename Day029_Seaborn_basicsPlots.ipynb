### Basic barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the tips dataset
tips = sns.load_dataset('tips')

# Barplot using sum instead of mean
sns.barplot(x='day', y='total_bill', data=tips, estimator=lambda x: np.sum(x))

plt.title('Total Bill by Day (Sum)')
plt.show()
```

### Grouped Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create a grouped bar chart: total bill by day, grouped by gender (sex)
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)

plt.title("Total Bill by Day (Grouped by Gender)")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.legend(title="Gender")
plt.show()
```

### Horizontal Bar Chart 

```python
# Horizontal bar chart: total bill by day
sns.barplot(y="day", x="total_bill", data=tips)

plt.title("Total Bill by Day (Horizontal)")
plt.xlabel("Total Bill")
plt.ylabel("Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a barplot of average total_bill by day
sns.barplot(x='day', y='total_bill', hue = 'sex',data=tips)

plt.title('Avg total bill for Day')
plt.show()

```

### countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')
# sns.countplot(x='day', hue='sex', data=tips)
sns.countplot(x='smoker', hue='sex', data=tips)
plt.title("Number of records each Day")
plt.show()
```

### boxplot 

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Boxplot of total bill per day
sns.boxplot(x = 'day',y = 'total_bill',data=tips,hue='sex')
# sns.boxplot(x='smoker', y='tip', data=tips)
plt.title("Total Bill Distribution by Day")
plt.show()
```

### violinplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Total Bill Distribution by Day")
plt.show()
```

### Stripplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load tips dataset
tips = sns.load_dataset("tips")

# Stripplot - total_bill by day
sns.stripplot(x="day", y="total_bill", data=tips,palette='Set2',hue='sex')
plt.title("Total Bill by Day (Stripplot)")
plt.show()
```

### Swarmplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load tips dataset
tips = sns.load_dataset("tips")

# Stripplot - total_bill by day
sns.swarmplot(x="day", y="total_bill", data=tips, palette="Set1")

plt.title("Boxplot + Swarmplot Overlay")
plt.show()
```

### Pointplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Pointplot: average total bill per day, with gender as hue
sns.pointplot(x="day", y="total_bill", data=tips,hue="sex",palette="Set1", capsize=0.1)
plt.title("Average Total Bill by Day and Gender")
plt.show()
```

### Catplot

```python
# Catplot using stripplot
sns.catplot(x="day", y="total_bill", kind="strip", data=tips)
plt.title("Catplot: Stripplot version")
plt.show()

# Catplot using boxplot with hue
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
plt.title("Catplot: Boxplot with hue")
plt.show()
```
