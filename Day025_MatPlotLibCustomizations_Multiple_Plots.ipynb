### Customization_Multiple_Plots

```python
# legend
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [10, 20, 25, 30, 35]
y2 = [5, 15, 20, 25, 30]
plt.plot(x,y1,label = 'Product A',color = 'blue')
plt.plot(x,y2,label = 'Product B',color = 'green')
plt.title("Sales Comparision")
plt.xlabel('Week')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

### customize the tick marks 

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [ i**2 for i in x]
y2 = [ i**3 for i in x]
plt.plot(x,y1,label='X Squared',color = 'blue')
plt.plot(x,y2,label='X Cubed',color = 'green')
plt.title('Squared and Cubed Grpah')
plt.xlabel('Numbers')
plt.ylabel('Squared and Cubed')
plt.grid(True)
plt.legend(loc = 'upper left')
plt.xlim(2,4)
plt.ylim(0,70)
plt.show()
```

### plt.xlim() and plt.ylim()

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [i**2 for i in x]
plt.plot(x,y)
plt.title('Custom Labels')
plt.xticks([1,2,3,4,5],['One','Two','Three','Four','Five'],rotation=30)
plt.yticks([0,5,10,15,20,25],rotation = 90)
plt.show()
```

### text() and annotate()

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

# Create the plot
# plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', color='purple')
plt.title("Adding Text and Annotations")
plt.xlabel("x")
plt.ylabel("x squared")

# Add simple text at position (2, 4)
plt.text(2, 4, "This is 2²", fontsize=10, color='blue')

# Annotate the point (5, 25) with an arrow from (3.5, 20)
plt.annotate("Peak here!", 
             xy=(5, 25), 
             xytext=(3.5, 20), 
             arrowprops=dict(facecolor='green', arrowstyle='->'),
             fontsize=10)

# Grid and layout
plt.grid(True)
plt.tight_layout()
plt.show()
```

### plt.subplots()

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

# Create 2×2 subplots
fig, ax = plt.subplots(2, 2,figsize=(10, 6))

# Top-left plot
ax[0, 0].plot(x, [i for i in x])
ax[0, 0].set_title("y = x")

# Top-right plot
ax[0, 1].plot(x, [i**2 for i in x])
ax[0, 1].set_title("y = x²")

# Bottom-left plot
ax[1, 0].plot(x, [i**3 for i in x])
ax[1, 0].set_title("y = x³")

# Bottom-right plot
ax[1, 1].plot(x, [i + 5 for i in x])
ax[1, 1].set_title("y = x + 5")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 15, 10, 20, 25]

plt.figure(figsize=(8, 5))

# First Line: solid, thick, semi-transparent
plt.plot(x, y1, label="Line 1 - Solid", 
         linewidth=3, linestyle='-', alpha=1, color='blue')

# Second Line: dashed, thin, more transparent
plt.plot(x, y2, label="Line 2 - Dashed", 
         linewidth=2, linestyle='--', alpha=0.5, color='green')

# Add labels and title
plt.title("Line Width, Alpha, and Linestyle Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

### linewidth, alpha, and linestyle
