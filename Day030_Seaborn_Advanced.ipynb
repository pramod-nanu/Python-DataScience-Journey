### Seaborn Distribution Plots

### Histogram with histplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")
print(tips.head())

# Create histogram with bins
sns.histplot(tips['total_bill'])
plt.show()
sns.histplot(tips['total_bill'],bins=3)
```

```python
sns.histplot(data=tips, x="total_bill", hue="sex", bins=20,multiple="dodge",color='yellow')
plt.title("Histogram by Sex")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram + KDE")
plt.show()
```

### kdeplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True,color='pink',bw_adjust=0.8)
plt.title("KDE Plot by Sex")
plt.show()
```

```python
sns.histplot(data=tips, x="tip", kde=True, hue="sex", element="step", stat="density", common_norm=False)
plt.title("Tip Distribution by Sex")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True, bins=20, color='skyblue', edgecolor='green')
plt.title("Styled Histogram + KDE")
plt.show()
```

### displot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
sns.displot(data=titanic, x='age', bins=20, col='sex', kde=True)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
sns.distplot(titanic['age'].dropna(), bins=15, kde=True, color='skyblue')
plt.title("Age Distribution of Titanic Passengers (distplot)")
plt.show()
```
