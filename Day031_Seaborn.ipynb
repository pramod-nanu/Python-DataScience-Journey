### More Plots


```python
### RugPlot
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.rugplot(x=tips['total_bill'],height = 0.2,color='blue')
plt.title("Rugplot of Total Bill")
plt.show()
```

```python
### Displot
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic data
titanic = sns.load_dataset("titanic")
# Basic histogram
sns.displot(data=titanic, x='age', hue='sex', bins=15, kind='hist', palette='Set2')
plt.title("Age Distribution of Titanic Passengers")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic data
titanic = sns.load_dataset("titanic")
# Basic histogram
sns.displot(data=titanic, x='age', kind='kde', fill=True, color='green', hue='sex')

plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

```python
### sns.kdeplot()
import seaborn as sns
import matplotlib.pyplot as plt
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
# KDE Plot of Age
sns.kdeplot(data=titanic, x='age',fill = True,color = 'blue')
plt.title('KDE Plot of Passenger Age')
plt.show()
```

```python
### sns.scatterplot()  Plotting relationships between two numeric variables
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='sex', style='class', size='survived')
plt.title('Scatterplot: Age vs Fare (class & survived)')
plt.show()

```

```python
### sns.lineplot()  For trends over time or sequence
# Sample dataset for line plot
flights = sns.load_dataset('flights')
print(flights.head())

# Lineplot of passengers per year
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title('Monthly Trends of Passengers')
plt.show()

```

```python
### relplot()  Flexible wrapper for scatter/line plots
# relplot as scatter
sns.relplot(data=titanic, x='age', y='fare', hue='sex', kind='scatter')
plt.title('relplot: Scatter')
plt.show()

# relplot as line
sns.relplot(data=flights, x='year', y='passengers', hue='month', kind='line')
plt.title('relplot: Line')
plt.show()

```
